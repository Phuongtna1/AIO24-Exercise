```python
def armstrong(x):
    lst = [int(i)**len(str(x)) for i in str(x)]
    if sum(lst) == x:
        return 1
    else:
        return 0
```

```python
test_case_1 = [130, 270, 153, 407, 177, 371, 1000, 1634, 370]
result = [x for x in test_case_1 if armstrong(x) == 1]
print(f"Các số Armstrong có trong list là: {result}")
```

## Section 4 - List and Branching


```python
def candies(lst, extra):
    result = []
    for i in range(len(lst)):
        lst[i] += extra
        result.append(lst[i] == max(lst))
    return result

test_case_1 = [2, 3, 5, 1, 3]
extraCandies1 = 3
print(candies(test_case_1, extraCandies1))
test_case_2 = [4, 2, 1, 1, 2]
extraCandies2 = 1
print(candies(test_case_2, extraCandies2))
test_case_3 = [12, 1, 12]
extraCandies3 = 10
print(candies(test_case_3, extraCandies3))
```

## Section 5 - Computing median for a list of numbers
 

```python
def median(lst):
    lst = sorted(lst)
    N = len(lst)
    if N%2 == 0:
        return (lst[int(N/2 - 1)] + lst[int(N/2)])/2
    else:
        return lst[int((N/2))]
```

```python
lst_data = list(range(1, 11))
lst_odd_filter = sorted([i for i in lst_data if i%2 == 1], reverse= True)

print(f"Cau 1: {lst_data}")
print(f"Cau 2: Median: {median(lst_data)}")
print(f"Cau 3: {lst_odd_filter}")
```

## Section 7 - Bag of Words (NLP)


```python
def Tokenization(doc, corpus):
    volcabulary = sorted(list(set(" ".join(corpus).split())))
    
    vector = []
    doc = doc.split()
    for w in volcabulary:
        count = [1 if i==w else 0 for i in doc]
        vector.append(sum(count))
    return vector, volcabulary
```

```python
corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"]
doc = "Tôi thích AI thích Toán"

vector, volcabulary = Tokenization(doc, corpus)
print(f"{doc}: {vector}")
print(f"Bag-of-Words: {volcabulary}")
```

## Section 9 - Interpolation for List (Time-series)


```python
def NNI(lst):    
    for i in range(len(lst)):
        if lst[i] is not None:
            neighbor = lst[i]
            break
    for i in range(len(lst)):
        if lst[i] is not None:
            neighbor = lst[i]
        else:
            lst[i] = neighbor
    return lst

lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]
print(NNI(lst_data))
            
```

## Section 11 - Matrix representation using List


```python
A = [[1, 2, 3], 
     [4, 5, 6], 
     [7, 8, 9]]
B = [[2, 4, 6], 
     [1, 3, 5], 
     [1, 0, 1]]

def sum_matrix(a, b):
    result = []
    for i in range(len(a)):
        row = []
        for j in range(len(b)):
            row.append(a[i][j] + b[i][j])
        result.append(row)
    return result

def sub_matrix(a, b):
    result = []
    for i in range(len(a)):
        row = []
        for j in range(len(b)):
            row.append(a[i][j] - b[i][j])
        result.append(row)
    return result

def dot_matrix(a, b):
    result = []
    for i in range(len(a)):
        row = []
        for j in range(len(b[0])):
            row.append(sum([a[i][k]*b[k][j] for k in range(len(a[i]))]))
        result.append(row)
    return result

print(f"Tong: {sum_matrix(A, B)}")
print(f"Hieu: {sub_matrix(A, B)}")
print(f"Dot Product: {dot_matrix(A, B)}")
```
